>>> #### Center City Locations

```python
def check_promo_month(row):
    month2str = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',              
                 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}
    try:
        months = (row['PromoInterval'] or '').split(',')
        if row['Promo2Open'] and month2str[row['Month']] in months:
            return 1
        else:
            return 0
    except Exception:
        return 0

def promo_cols(df):
    # Months since Promo2 was open
    df['Promo2Open'] = 12 * (df.Year - df.Promo2SinceYear) +  (df.WeekOfYear - df.Promo2SinceWeek)*7/30.5
    df['Promo2Open'] = df['Promo2Open'].map(lambda x: 0 if x < 0 else x).fillna(0) * df['Promo2']
    # Whether a new round of promotions was started in the current month
    df['IsPromo2Month'] = df.apply(check_promo_month, axis=1) * df['Promo2']
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import sys
sys.path.append('../')
from scripts.cleaning import *
```

```python
merged_train = pd.read_csv('../docs/merged_train.csv')
merged_test = pd.read_csv('../docs/merged_test.csv')
pd.set_option('display.max_columns', None)
merged_train.head(1)
```

```python
# Find common columns
common_columns = merged_train.columns.intersection(merged_test.columns)
selected_columns = list(common_columns) + ['Sales']

# Extract common columns from merged_train
merged_train = merged_train[selected_columns]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pickle


# Load your preprocessed data
X_train = preprocess_data(merged_train.copy())
y_train = X_train.pop('Sales')

# Preprocess the test data
X_test = preprocess_data(merged_test)

numeric_features = X_train.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Define the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features)  # Scale all numeric features
    ],
    remainder='passthrough'  # Leave other columns untouched
)

# Define the pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),  # Apply preprocessing
    ('regressor', RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1))  # Random Forest Regressor
])
```

```python
X_train.columns.unique()
```

```python
pipeline.fit(X_train, y_train)

# Save the pipeline
with open('model_pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline, f)

with open('scaler.pkl', 'wb') as f:
    pickle.dump(preprocessor.named_transformers_['num'], f)

# Make predictions on the training data for evaluation
y_train_pred = pipeline.predict(X_train)

# Evaluate the model
mae = mean_absolute_error(y_train, y_train_pred)
mse = mean_squared_error(y_train, y_train_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_train, y_train_pred)

print(f'Training Set Evaluation:')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R): {r2}')
```

>> ## Confidence Intervals By Bootstrapping

```python
from sklearn.utils import resample

# Function to perform bootstrapping and calculate confidence intervals
def bootstrap_predictions(model, X, y, n_iterations=10, alpha=0.05):
    predictions = []
    
    for _ in range(n_iterations):
        # Resample the data
        X_resampled, y_resampled = resample(X, y)
        model.fit(X_resampled, y_resampled)
        pred = model.predict(X)
        predictions.append(pred)
    
    # Convert to DataFrame for easier handling
    predictions = np.array(predictions)
    
    # Calculate lower and upper percentiles for confidence intervals
    lower_percentile = np.percentile(predictions, 100 * (alpha / 2), axis=0)
    upper_percentile = np.percentile(predictions, 100 * (1 - alpha / 2), axis=0)
    
    return lower_percentile, upper_percentile

# Estimate confidence intervals for test predictions
lower_ci, upper_ci = bootstrap_predictions(pipeline.named_steps['regressor'], X_test, y_test_pred)

# Save predictions along with confidence intervals to a DataFrame
predictions_df = pd.DataFrame({
    'Predicted_Sales': y_test_pred,
    'Lower_CI': lower_ci,
    'Upper_CI': upper_ci
})

predictions_df.to_csv('../docs/predictions_with_ci.csv', index=False)

print(predictions_df.head())  # Display the predictions along with confidence intervals
```

```python
from statsmodels.tsa.stattools import adfuller

def check_stationarity(timeseries):
    result = adfuller(timeseries['Sales'])
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] <= 0.05:
        print("The time series is stationary.")
    else:
        print("The time series is not stationary.")
        
check_stationarity(merged_train.sample(n=5000))
```

```python
plt.figure(figsize=(12, 6))
plot_acf(merged_train['Sales'].sample(5000), lags=30)
plt.title('Autocorrelation Function (ACF)')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(merged_train['Sales'].sample(5000), lags=30)
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 30  # Use past 30 days to predict the next day
data_values = merged_train['Sales'].values.reshape(-1, 1)

# Scale the data
scaler = MinMaxScaler(feature_range=(-1, 1))
data_scaled = scaler.fit_transform(data_values)

# Create the dataset
X, y = create_dataset(data_scaled, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)  # Reshape for LSTM input
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))  # Dropout for regularization
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(1))  # Output layer

model.compile(optimizer='adam', loss='mean_squared_error')
```

>> ## Train the model

```python
# This may result in underfitting, but my pc can't handle large number of epochs. Ideally you would want to go through the dataset atleast 50 times.
history = model.fit(X, y, epochs=2, batch_size=32, verbose=1)
```
